```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.models import Model
import numpy as np

# Load ResNet50 without the top layer
resnet_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')

def extract_image_features(image_paths):
    features = []
    for path in image_paths:
        try:
            img = load_img(path, target_size=(224, 224))
            img_array = img_to_array(img)
            img_array = preprocess_input(img_array)  # ResNet50 preprocessing
            img_array = np.expand_dims(img_array, axis=0)
            feature = resnet_model.predict(img_array)
            features.append(feature.flatten())
        except Exception as e:
            print(f"Error loading image: {path}")
            features.append(np.zeros((2048,)))  # Placeholder for missing images
    return np.array(features)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Tokenization
tokenizer = Tokenizer(num_words=10000, oov_token="<UNK>")
tokenizer.fit_on_texts(df_train['tweet_text'])

max_len = 50  # Adjust based on analysis of tweet length
X_text_train = pad_sequences(tokenizer.texts_to_sequences(df_train['tweet_text']), maxlen=max_len, padding='post')
X_text_val = pad_sequences(tokenizer.texts_to_sequences(df_val['tweet_text']), maxlen=max_len, padding='post')
X_text_test = pad_sequences(tokenizer.texts_to_sequences(df_test['tweet_text']), maxlen=max_len, padding='post')
```

```python
from tensorflow.keras.layers import Embedding, LSTM, Dropout, Dense, Input
from tensorflow.keras.models import Model

vocab_size = len(tokenizer.word_index) + 1  # Include padding and OOV tokens

# Text Model
input_text = Input(shape=(max_len,))
embedding = Embedding(input_dim=vocab_size, output_dim=300)(input_text)
lstm = LSTM(256, return_sequences=False)(embedding)
dropout = Dropout(0.3)(lstm)
text_output = Dense(128, activation='relu')(dropout)
```

```python
from tensorflow.keras.layers import Concatenate

# Image Model
input_image = Input(shape=(2048,))  # ResNet50 output size
image_dense = Dense(128, activation='relu')(input_image)

# Combine Models
combined = Concatenate()([text_output, image_dense])
dropout_combined = Dropout(0.3)(combined)
output = Dense(len(df_train['label_text'].unique()), activation='softmax')(dropout_combined)

# Final Model
model = Model(inputs=[input_text, input_image], outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
def get_image_path(image_id, base_path="/content/CrisisMMD_v2.0/data_image"):
    for root, dirs, files in os.walk(base_path):
        if image_id + ".jpg" in files:
            return os.path.join(root, image_id + ".jpg")
    return None
```

```python
train_image_paths = df_train['image_id'].apply(lambda x: get_image_path(x))
val_image_paths = df_val['image_id'].apply(lambda x: get_image_path(x))
test_image_paths = df_test['image_id'].apply(lambda x: get_image_path(x))

# Extract features
X_img_train = extract_image_features(train_image_paths)
X_img_val = extract_image_features(val_image_paths)
X_img_test = extract_image_features(test_image_paths)
```

```python
model.fit(
    [X_text_train, X_img_train],
    pd.get_dummies(df_train['label_text']).values,
    validation_data=([X_text_val, X_img_val], pd.get_dummies(df_val['label_text']).values),
    epochs=10,
    batch_size=32
)
```
